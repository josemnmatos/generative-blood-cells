```python
from utils import load_medMNIST_data
```

```python
dataloader, full_dataset = load_medMNIST_data()
```

```python
# get 49 samples from the dataloader

samples = []
for i, (images, labels) in enumerate(dataloader):
    if i >= 49:
        break
    for j in range(len(images)):
        samples.append((images[j], labels[j]))
        
        
```

```python
# display generated images
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import torchvision.utils as vutils
import torch

image_tensors = [img for img, label in samples[:49]]  # 49 images for a 7x7 grid
images_tensor = torch.stack(image_tensors)  # shape: [49, 3, 28, 28]

def show_generated_images(images):
    # Create a grid with WHITE padding (pad_value=1.0)
    grid = vutils.make_grid(images, nrow=7, padding=2, normalize=True, pad_value=1.0)
    plt.figure(figsize=(8,8))
    plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
    plt.axis('off')
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# Show the images
show_generated_images(images_tensor)
```
